### 5.7 Evaluating the Necessity of Feature Scaling

Data normalization can improve the performance of some machine learning algorithms, particularly those that rely on numerical distances or gradients, but not all algorithms are equally affected. Here is an analysis of how normalization impacts various algorithms:

- **KNN (k-Nearest Neighbors)**: KNN uses Euclidean distance to calculate similarity between data points. When variables have different scales, variables with larger values can disproportionately affect the distance. Normalization ensures that all variables contribute equally to the distance measure, potentially improving KNN's accuracy.  
- **Logistic Regression**: Logistic regression employs the sigmoid function, which is sensitive to the scale of input variables. If some variables have significantly larger values than others, they may dominate the sigmoid function, leading to slow convergence or poor performance. Normalization can help mitigate this issue by ensuring all variables are on a similar scale.  
- **SVM (Support Vector Machine)**: SVM also relies on Euclidean distance to find the optimal hyperplane. Like KNN, variables with different scales can disproportionately influence the distance. Normalization can improve SVM performance, especially for datasets with features on different scales.  
- **Random Forests**: Similar to decision trees, random forests are relatively insensitive to normalization as they consist of multiple decision trees trained on different data subsets.  
- **Naive Bayes**: Naive Bayes is typically unaffected by normalization. This algorithm calculates probabilities based on the likelihood of each feature value given a class, assuming independence between features. Since it operates directly on the probability distribution, the absolute scale of the variables does not influence the calculations. However, extreme differences in scale might still indirectly affect numerical stability when computing probabilities in some implementations.  

In summary, normalization is a critical preprocessing step for algorithms that use distance metrics or gradient-based optimization methods. It ensures fair contribution from all features, improves convergence, and often enhances overall performance. On the other hand, algorithms like random forests, and Naive Bayes are inherently robust to differences in feature scale and generally do not require normalization.

### 5.9 Visualizing the Data in 2D

We will use Dimensionality Reduction techniques to visualize the data in 2D. We will use **PCA (Principal Component Analysis)** to reduce the dimensionality of the data to 2 components and plot the data points in a 2D space.

The **PCA** algorithm will help us understand the distribution of the data and the separability of the classes.

The **PCA** works by finding the directions (principal components) along which the variance of the data is maximized. These principal components are orthogonal to each other, and the first principal component captures the most variance in the data. Each of the subsequent principal components captures the remaining variance in decreasing order. By reducing the dimensionality of the data to 2 components, we can visualize the data in a 2D space.

````python
def reduce_dim_and_print(n_dimns=2, show_index=False):
    # Reduce dimensions to n_dimns
    pca = PCA(n_components=n_dimns)
    numeric_cols = df_scaled.select_dtypes(include=['number']).columns
    df_pca = pca.fit_transform(df_scaled[numeric_cols])

    # Plot the clusters
    plt.figure(figsize=(8, 6))
    plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df_yields_scaled['Yield'])
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('Clusters after PCA')
    plt.colorbar(label='Label')
    # Add index as text annotation for each point
    if show_index:
        for i, txt in enumerate(df_scaled.index):
            plt.annotate(txt, (df_pca[i, 0], df_pca[i, 1]),
                         textcoords="offset points", xytext=(5,5), ha='center', fontsize=8)  # Adjust text properties

    plt.show()

reduce_dim_and_print()
````

### 6.1 Avaliação dos Modelos

A função `evaluate_model` realiza a avaliação de um modelo de machine learning utilizando métricas de erro e exibe uma comparação entre os valores reais e previstos. A função segue os seguintes passos:

1. **Predição dos Valores**: Utiliza o modelo fornecido para prever os valores de `X_test` e armazena os resultados em `y_pred`.
    ```python
    y_pred = model.predict(X_test)
    ```

2. **Cálculo das Métricas de Erro**:
    - **Erro Quadrático Médio (MSE)**: Mede a média dos quadrados dos erros, ou seja, a diferença média ao quadrado entre os valores reais e previstos.
    - **Erro Absoluto Médio (MAE)**: Mede a média das diferenças absolutas entre os valores reais e previstos.
    - **Coeficiente de Determinação (R²)**: Indica a proporção da variância dos dados que é explicada pelo modelo.
    ```python
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    ```

3. **Exibição de Exemplos de Comparação**: Imprime uma comparação entre os primeiros 10 valores reais e previstos.
    ```python
    print("Some Samples to compare:")
    print("\nActual\tPredicted")
    for i in range(10):
        print(f"{y_test.iloc[i]}\t{y_pred[i]}")
    ```

4. **Exibição das Métricas**: Imprime as métricas calculadas (MSE, MAE e R²) com três casas decimais.
    ```python
    print("\nMetrics")
    print(
        f"\nMean Squared Error: {mse:.3f}\t Mean Absolute Error: {mae:.3f}\t R^2 Score: {r2:.3f}"
    )
    ```

5. **Retorno das Métricas**: Retorna as métricas calculadas para uso posterior.
    ```python
    return mse, mae, r2
    ```

Essa função é essencial para entender o desempenho do modelo em termos de precisão e erro, permitindo ajustes e melhorias no modelo de machine learning.

````python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)


    print("Some Samples to compare:")
    print("\nActual\tPredicted")
    for i in range(10):
        print(f"{y_test.iloc[i]}\t{y_pred[i]}")
    
    print("\nMetrics")
    print(
        f"\nMean Squared Error: {mse:.3f}\t Mean Absolute Error: {mae:.3f}\t R^2 Score: {r2:.3f}"
    )

    return mse, mae, r2
````

### 6.3.1 Regressão K-Nearest Neighbors (KNR)

Um algoritmo baseado na proximidade entre os pontos de dados para fazer previsões. Ele assume que valores próximos no espaço de recursos terão saídas semelhantes.

**Como funciona:**
- **Cálculo de distância**: Para cada nova amostra, o algoritmo calcula a distância entre essa amostra e todos os pontos do conjunto de treinamento (distâncias comuns incluem Euclidiana, Manhattan e Minkowski).
- **Seleção dos vizinhos mais próximos**: Após calcular as distâncias, seleciona-se um número `K` de vizinhos mais próximos da amostra a ser prevista.
- **Média dos vizinhos**: A previsão é feita com base na média (ou outra função estatística) dos valores dos `K` vizinhos mais próximos.

**Vantagens:**
- Simples de entender e implementar.
- Funciona bem para pequenos conjuntos de dados sem padrões complexos.
- Não assume nenhuma hipótese sobre a distribuição dos dados.

**Desvantagens:**
- Pode ser computacionalmente caro para grandes conjuntos de dados.
- Sensível a ruídos e outliers.
- A escolha do valor de `K` pode afetar significativamente o desempenho do modelo.

````python
def train_knr(X_train, y_train, X_test, y_test):
    # Initialize the KNN classifier with k=5 (you can adjust this value)
    knr = KNeighborsRegressor(n_neighbors=5)

    # Train the model
    knr.fit(X_train, y_train)

    # Evaluate the model
    evaluate_model(knr, X_test, y_test)
    
    return knr

train_knr(X_train, y_train)
````

### 6.3.2 Regressão de Máquina de Vetores de Suporte (SVR)

Um modelo baseado no conceito de Máquinas de Vetores de Suporte (SVM), adaptado para problemas de regressão. Ele busca encontrar uma função que melhor se ajuste aos dados, permitindo uma margem de erro dentro de um intervalo especificado.

**Como funciona:**
- **Definição da margem de tolerância (\(\epsilon\))**: O modelo estabelece uma margem dentro da qual os erros são considerados aceitáveis.
- **Maximização da margem**: Ao invés de minimizar diretamente os erros, o SVR busca encontrar uma função que mantenha a maioria dos pontos dentro da margem de erro definida.
- **Uso de kernels**: O SVR pode aplicar funções de kernel para transformar os dados em espaços de maior dimensão, permitindo capturar padrões não lineares.
- **Minimização do erro com vetores de suporte**: Apenas os pontos fora da margem (\(\epsilon\)-insensitive loss) contribuem para a otimização do modelo, reduzindo a complexidade computacional.

**Vantagens:**
- Pode capturar relações não lineares entre variáveis por meio do uso de kernels.
- Lida bem com dados de alta dimensão.
- Robusto a outliers devido ao controle da margem de erro.

**Desvantagens:**
- Pode ser computacionalmente caro para grandes conjuntos de dados.
- A escolha do kernel e dos hiperparâmetros (como \(\epsilon\) e \(C\)) pode ser difícil e impacta fortemente o desempenho.
- Pode não ser intuitivo de interpretar, especialmente com kernels não lineares.

````python
def train_svr(X_train, y_train, X_test, y_test):
    # Initialize the SVM classifier
    svr = SVR()

    # Train the model
    svr.fit(X_train, y_train)
    
    # Evaluate the model
    evaluate_model(svr, X_test, y_test)
    
    return svr
    
train_svr(X_train, y_train)
````

### 6.3.3 Regressão de Floresta Aleatória (RF)

A Regressão de Floresta Aleatória é um método baseado em múltiplas árvores de decisão, combinadas para melhorar a precisão e reduzir o risco de overfitting. Ele pertence à classe dos métodos de ensemble, onde vários modelos fracos se unem para formar um modelo forte.

**Como funciona:**
- **Criação de múltiplas árvores de decisão**: O modelo constrói várias árvores de decisão em subconjuntos diferentes do conjunto de treinamento.
- **Bagging (Bootstrap Aggregating)**: Cada árvore recebe um subconjunto aleatório dos dados, escolhido com reposição, garantindo diversidade entre as árvores.
- **Cálculo da média das previsões**: Para prever um novo valor, cada árvore faz uma estimativa, e a previsão final é a média dessas estimativas.

**Vantagens:**
- Reduz o overfitting em comparação com uma única árvore de decisão.
- Funciona bem com dados de alta dimensionalidade e relações não lineares.
- Robusto a outliers e dados ruidosos devido à combinação de várias árvores.

**Desvantagens:**
- Pode ser computacionalmente intensivo, especialmente para grandes conjuntos de dados.
- Difícil de interpretar devido à natureza de múltiplas árvores.
- Requer mais memória para armazenar várias árvores do modelo.

````python
def train_rf(X_train, y_train, X_test, y_test):
    # Initialize the Random Forest classifier
    rf = RandomForestRegressor(random_state=42)

    # Train the model
    rf.fit(X_train, y_train)

    evaluate_model(rf, X_test, y_test)
    
    return rf

train_rf(X_train, y_train)
````

### 6.3.4 Regressão Linear  

A Regressão Linear é um dos modelos mais simples e amplamente utilizados em machine learning. Ela assume uma relação linear entre a variável de saída (dependente) e uma ou mais variáveis de entrada (independentes).  

**Como funciona:**  
- **Definição da equação linear**: O modelo ajusta uma equação da forma:  
  ```
  y = b0 + b1*x1 + b2*x2 + ... + bn*xn
  ```
  onde `y` é o valor previsto, `x1, x2, ..., xn` são as variáveis de entrada e `b0, b1, ..., bn` são os coeficientes que o modelo aprende durante o treinamento.  
- **Ajuste dos coeficientes**: O modelo encontra os valores de `b0, b1, ..., bn` que minimizam a diferença entre as previsões e os valores reais, geralmente utilizando o método dos mínimos quadrados.  
- **Interpretação dos coeficientes**: Cada coeficiente `b1, b2, ..., bn` representa a influência de sua respectiva variável de entrada na previsão do modelo.  

**Vantagens:**  
- Simples de implementar e interpretar.  
- Rápido para treinar, mesmo em conjuntos de dados grandes.  
- Funciona bem quando há uma relação linear entre as variáveis.  

**Desvantagens:**  
- Não captura relações não lineares entre as variáveis.  
- Sensível a outliers, que podem distorcer os coeficientes.  
- Pode apresentar problemas se houver alta correlação entre as variáveis de entrada (multicolinearidade).  

````python
def train_dtr(X_train, y_train, X_test, y_test):
    # Initialize the Decision Tree classifier
    dtr = DecisionTreeRegressor(random_state=42)

    # Train the model
    dtr.fit(X_train, y_train)

    evaluate_model(dtr, X_test, y_test)
    
    return dtr

train_dtr(X_train, y_train)
````

### 6.3.5 Regressão de Árvore de Decisão  

A Regressão de Árvore de Decisão é um modelo baseado em árvores de decisão que particiona o espaço dos dados em regiões distintas, ajustando valores constantes dentro de cada uma delas.  

**Como funciona:**  
- **Divisão recursiva dos dados**: O modelo divide os dados iterativamente em diferentes regiões, escolhendo os pontos de divisão que minimizam a variabilidade dentro de cada subconjunto.  
- **Critério de divisão**: O ponto de corte é selecionado para minimizar uma métrica como o erro quadrático médio (MSE).  
- **Previsão com valores médios**: Para uma nova amostra, o modelo percorre a árvore até um nó terminal e retorna a média dos valores observados nesse nó.  

**Vantagens:**  
- Fácil de interpretar e visualizar.  
- Captura relações não lineares entre as variáveis.  
- Funciona bem com dados mistos (numéricos e categóricos).  

**Desvantagens:**  
- Propenso a overfitting, especialmente se a árvore for muito profunda.  
- Sensível a pequenas variações nos dados, podendo gerar árvores muito diferentes (instáveis).  
- Pode não generalizar bem para novos dados se não for podado corretamente.  

````python
def train_lr(X_train, y_train, X_test, y_test):
    # Initialize the Linear Regression model
    lr = LinearRegression()

    # Train the model
    lr.fit(X_train, y_train)

    # Evaluate the model
    evaluate_model(lr, X_test, y_test)
    
    return lr

train_lr(X_train, y_train, X_test, y_test)
````
