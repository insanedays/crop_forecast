### 5.7 Evaluating the Necessity of Feature Scaling

Data normalization can improve the performance of some machine learning algorithms, particularly those that rely on numerical distances or gradients, but not all algorithms are equally affected. Here is an analysis of how normalization impacts various algorithms:

- **KNN (k-Nearest Neighbors)**: KNN uses Euclidean distance to calculate similarity between data points. When variables have different scales, variables with larger values can disproportionately affect the distance. Normalization ensures that all variables contribute equally to the distance measure, potentially improving KNN's accuracy.  
- **Logistic Regression**: Logistic regression employs the sigmoid function, which is sensitive to the scale of input variables. If some variables have significantly larger values than others, they may dominate the sigmoid function, leading to slow convergence or poor performance. Normalization can help mitigate this issue by ensuring all variables are on a similar scale.  
- **SVM (Support Vector Machine)**: SVM also relies on Euclidean distance to find the optimal hyperplane. Like KNN, variables with different scales can disproportionately influence the distance. Normalization can improve SVM performance, especially for datasets with features on different scales.  
- **Random Forests**: Similar to decision trees, random forests are relatively insensitive to normalization as they consist of multiple decision trees trained on different data subsets.  
- **Naive Bayes**: Naive Bayes is typically unaffected by normalization. This algorithm calculates probabilities based on the likelihood of each feature value given a class, assuming independence between features. Since it operates directly on the probability distribution, the absolute scale of the variables does not influence the calculations. However, extreme differences in scale might still indirectly affect numerical stability when computing probabilities in some implementations.  

In summary, normalization is a critical preprocessing step for algorithms that use distance metrics or gradient-based optimization methods. It ensures fair contribution from all features, improves convergence, and often enhances overall performance. On the other hand, algorithms like random forests, and Naive Bayes are inherently robust to differences in feature scale and generally do not require normalization.

### 5.8 Applying Feature Scaling

````python
df_features = df.drop('label', axis=1)

# Identify numeric features
numeric_features = df_features.select_dtypes(include=['number']).columns

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Copy
df_scaled = df.copy()

# Fit and transform the numeric features
df_scaled[numeric_features] = pd.DataFrame(scaler.fit_transform(df[numeric_features]), columns=numeric_features)
df_scaled.head()
````

### 5.9 Visualizing the Data in 2D

We will use Dimensionality Reduction techniques to visualize the data in 2D. We will use **PCA (Principal Component Analysis)** to reduce the dimensionality of the data to 2 components and plot the data points in a 2D space.

The **PCA** algorithm will help us understand the distribution of the data and the separability of the classes.

The **PCA** works by finding the directions (principal components) along which the variance of the data is maximized. These principal components are orthogonal to each other, and the first principal component captures the most variance in the data. Each of the subsequent principal components captures the remaining variance in decreasing order. By reducing the dimensionality of the data to 2 components, we can visualize the data in a 2D space.

````python
def reduce_dim_and_print(n_dimns=2, show_index=False):
    # Reduce dimensions to n_dimns
    pca = PCA(n_components=n_dimns)
    df_pca = pca.fit_transform(df_scaled.drop('label', axis=1))

    # Plot the clusters
    plt.figure(figsize=(8, 6))
    plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df_scaled['label'])
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title('Clusters after PCA')
    plt.colorbar(label='Label')
    # Add index as text annotation for each point
    if show_index:
        for i, txt in enumerate(df_scaled.index):
            plt.annotate(txt, (df_pca[i, 0], df_pca[i, 1]),
                         textcoords="offset points", xytext=(5,5), ha='center', fontsize=8)  # Adjust text properties

    plt.show()

reduce_dim_and_print()

# Print counts for each cluster
df_scaled['label'].value_counts().sort_index()
````

### 6.1 Dataset Split

We will split the dataset into training and testing sets using the `train_test_split` function from `scikit-learn`. We will use a 70-30 split, where 70% of the data will be used for training the model, and 30% will be used for testing the model.

The `stratify` parameter will be used to ensure that the class distribution is preserved in the training and testing sets. This is important to ensure that the model is trained on a representative sample of the data and can generalize well to unseen data.

````python
# Assuming df_scaled is your DataFrame with the 'cluster' column
X = df_scaled.drop('label', axis=1)
y = df_scaled['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
````

#### 6.2.1 K-Nearest Neighbors (KNN)

An algorithm based on proximity between points for classification.

**How it works:**
- **Distance calculation**: Measures the distance between the new point and all existing points.  
- **Neighbor selection**: Identifies the K nearest neighbors.  
- **Voting**: Classifies based on the majority of the neighbors.  

**Advantages:**
- Simple to implement.  
- Does not require training.  
- Adaptable to new data.  

**Disadvantages:**
- Computationally intensive for large datasets.  
- Sensitive to the choice of the K value.  
- Requires data normalization.  

````python
# Initialize the KNN classifier with k=3 (you can adjust this value)
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model
knn.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred, digits=4))
print(confusion_matrix(y_test, y_pred))
````

#### 6.2.2 Support Vector Machine (SVM)

An algorithm based on proximity between points for classification.

**How it works:**
- **Distance calculation**: Measures the distance between the new point and all existing points.  
- **Neighbor selection**: Identifies the K nearest neighbors.  
- **Voting**: Classifies based on the majority of the neighbors.  

**Advantages:**

#### 6.2.3 Random Forest

An evolution of the decision tree that combines multiple trees to improve predictive capability.  

**How it works:**  
- **Creation of multiple trees**: Trains several trees using random subsets of the data.  
- **Aggregation**: Combines the predictions from all trees through voting or averaging.  

**Advantages:**  
- High accuracy.  
- Robust against overfitting.  
- Good performance with high-dimensional data.  

**Disadvantages:**  
- Lower interpretability.  
- High computational cost.  

````python
# Initialize the Random Forest classifier
rf = RandomForestClassifier(random_state=42)

# Train the model
rf.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred_rf, digits=4))
print(confusion_matrix(y_test, y_pred_rf))
````

#### 6.2.4 Naive Bayes

A probabilistic algorithm based on Bayes' theorem with an assumption of feature independence.  

**How it works:**  
- **Probability calculation**: Computes the probability of each class given the input features.  
- **Feature independence assumption**: Assumes all features are independent, simplifying probability calculations.  
- **Classification**: Assigns the class with the highest posterior probability to the data point.  

**Advantages:**  
- Simple and fast to train.  
- Works well with small datasets.  
- Effective for high-dimensional data.  

**Disadvantages:**  
- Assumes feature independence, which may not hold in real-world datasets.  
- May perform poorly if features are highly correlated.  
- Sensitive to imbalanced datasets.  

````python
# Initialize the Naive Bayes classifier
nb = GaussianNB()

# Train the model
nb.fit(X_train, y_train)

# Make predictions on the test set
y_pred_nb = nb.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred_nb, digits=4))
print(confusion_matrix(y_test, y_pred_nb))
````

#### 6.2.5 Logistic Regression

````python
# Initialize the Logistic Regression classifier
lr = LogisticRegression(random_state=42, max_iter=1000)

# Train the model
lr.fit(X_train, y_train)

# Make predictions on the test set
y_pred_lr = lr.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred_lr, digits=4))
print(confusion_matrix(y_test, y_pred_lr))
````

### 6.3 Training Hyper-Parameter Optimized Models

#### 6.3.1 K-Nearest Neighbors (KNN)

````python
# Define the parameter grid
param_grid = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan", "minkowski"],
    "algorithm": ["ball_tree", "kd_tree", "brute"],
    "leaf_size": [10, 20, 30, 40, 50],
}

# Initialize the KNN classifier
knn_g = KNeighborsClassifier()

# Initialize GridSearchCV with the KNN classifier and the parameter grid
grid_search = GridSearchCV(
    estimator=knn_g, param_grid=param_grid, cv=5, scoring="accuracy"
)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters and the best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best Parameters: {best_params}")
print(f"Best Cross-Validation Score: {best_score:.4f}")

# Use the best estimator to make predictions on the test set
best_knn = grid_search.best_estimator_
y_pred_best_knn = best_knn.predict(X_test)

# Evaluate the optimized model
print(classification_report(y_test, y_pred_best_knn, digits=4))
print(confusion_matrix(y_test, y_pred_best_knn))
````
